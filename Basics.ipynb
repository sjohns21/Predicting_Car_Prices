```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
def knn_train_test(df, train_col, target_col):
    #randomize rows
    split_index = round(cars.shape[0] * .75)
    #train test split
    train_df = df.iloc[:split_index]
    test_df = df.iloc[split_index:]
    #create, fit, and predict with model
    knn = KNeighborsRegressor()
    knn.fit(train_df[[train_col]],train_df[target_col])
    preds = knn.predict(test_df[[train_col]])
    #calculate and return error
    rmse = mean_squared_error(test_df[target_col], preds) ** .5
    return rmse
```

```python
def knn_train_test(df, train_col, target_col, k):
    #randomize rows
    split_index = round(cars.shape[0] * .75)
    #train test split
    train_df = df.iloc[:split_index]
    test_df = df.iloc[split_index:]
    #create, fit, and predict with model
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(train_df[[train_col]],train_df[target_col])
    preds = knn.predict(test_df[[train_col]])
    #calculate and return error
    rmse = mean_squared_error(test_df[target_col], preds) ** .5
    return rmse
```

# multivariate model

```python
def knn_train_test(df, train_cols, target_col, k):
    #randomize rows
    split_index = round(cars.shape[0] * .75)
    #train test split
    train_df = df.iloc[:split_index]
    test_df = df.iloc[split_index:]
    #create, fit, and predict with model
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(train_df[train_cols],train_df[target_col])
    preds = knn.predict(test_df[train_cols])
    #calculate and return error
    rmse = mean_squared_error(test_df[target_col], preds) ** .5
    return rmse
```

```python
knn_train_test(cars, ranked_cols, 'price', 5)
```

```python
d={}
for i in range(2,6):
    d[i]=knn_train_test(cars, ranked_cols[:i], 'price', 5)
```

```python
d
```

```python
pd.Series(d)
```

```python
pd.Series(d).sort_values()
```

top 3, top 4, and top 2 features were the highest performing

```python
final_df = pd.DataFrame(index=range(1,26), columns=[2,3,4])
```

```python
final_df.head()
```

```python
for top_n_cols in range(2,5):
    for k_val in range(1,26):
        final_df.loc[k_val, top_n_cols] = knn_train_test(cars, ranked_cols[:top_n_cols], 'price', k_val)
        
```

```python
final_df
```

```python
final_df.plot()
plt.xlabel('k-value')
plt.ylabel('rmse')
plt.legend(title='using top n features')
```

as we can see, our model's error is minimized when using the top few features and using k values of 1-3.
